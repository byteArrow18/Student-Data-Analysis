# Task 1: Data Preparation


```python
# Start from here to write your code for Task 1. Please also feel free to add more cells in this ipynb file. 
import pandas as pd

data = pd.read_csv('cleaned_A1data.csv')
```

```python

data.columns = data.columns.str.strip()

percentage_columns = ['Total', 'Rural (Residence)', 'Urban (Residence)',
                      'Poorest (Wealth quintile)', 'Richest (Wealth quintile)']
data[percentage_columns] = data[percentage_columns].replace({'%': ''}, regex=True).astype(float)
print(data)
```

# Task 2: Data Exploration

## Task 2.1 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
data.boxplot(column='Total', by='Income Group', vert=False)
plt.xlabel('Percentage with Internet Connection (Urban)')
plt.ylabel('Income Group')
plt.title('Boxplot of Internet Connection by Income Group')
plt.show()
```

## Task 2.2 

```python

median_rural = data.groupby('Region')['Rural (Residence)'].median()
median_urban = data.groupby('Region')['Urban (Residence)'].median()

print("\nMedian Percentage with Internet Connection (Rural):")
print(median_rural)
print("\nMedian Percentage with Internet Connection (Urban):")
print(median_urban)

topRural = data.nlargest(10, 'Rural (Residence)')
print("\nTop 10 Countries with Highest Percentage (Rural):\n\n", topRural)

topUrban = data.nlargest(10, 'Urban (Residence)')
print("\nTop 10 Countries with Highest Percentage (Urban):\n", topUrban)
```

## Task 2.3 

```python
upperm_data = data[data['Income Group'] == 'Upper middle income (UM)']

poorest_mean = upperm_data['Poorest (Wealth quintile)'].mean()
richest_mean = upperm_data['Richest (Wealth quintile)'].mean()

poorest_median = upperm_data['Poorest (Wealth quintile)'].median()
richest_median = upperm_data['Richest (Wealth quintile)'].median()

poorest_std = upperm_data['Poorest (Wealth quintile)'].std()
richest_std = upperm_data['Richest (Wealth quintile)'].std()


print("\nUpper Middle Income Group:")
print(f"Mean Percentage for Poorest: {poorest_mean:.2f}%")
print(f"Mean Percentage for Richest: {richest_mean:.2f}%")

print(f"Median Percentage for Poorest: {poorest_median:.2f}%")
print(f"Median Percentage for Richest: {richest_median:.2f}%")

print(f"Standard Deviation for Poorest: {poorest_std:.2f}")
print(f"Standard Deviation for Richest: {richest_std:.2f}")
```
